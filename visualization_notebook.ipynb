```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
prices_d1 = pd.read_csv('training_capsule/prices_round_0_day_-1.csv', sep=';')
prices_d2 = pd.read_csv('training_capsule/prices_round_0_day_-2.csv', sep=';')
trades_d1 = pd.read_csv('training_capsule/trades_round_0_day_-1.csv', sep=';')
trades_d2 = pd.read_csv('training_capsule/trades_round_0_day_-2.csv', sep=';')
```

```python
emeralds_prices_d1 = prices_d1[prices_d1['product'] == 'EMERALDS'].reset_index(drop=True)
tomatoes_prices_d1 = prices_d1[prices_d1['product'] == 'TOMATOES'].reset_index(drop=True)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8))

ax1.plot(emeralds_prices_d1['timestamp'], emeralds_prices_d1['mid_price'], linewidth=0.8)
ax1.set_title('EMERALDS Mid Price D1')
ax1.set_ylabel('Price')

ax2.plot(tomatoes_prices_d1['timestamp'], tomatoes_prices_d1['mid_price'], linewidth=0.8, color='red')
ax2.set_title('TOMATOES Mid Price D1')
ax2.set_ylabel('Price')
ax2.set_xlabel('Timestamp')

print(emeralds_prices_d1['mid_price'].min(), emeralds_prices_d1['mid_price'].max())
print(emeralds_prices_d1['mid_price'].unique()[:20])

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8))

ax1.plot(emeralds_trades_d1['timestamp'], emeralds_trades_d1['price'], linewidth=0.8)
ax1.set_title('EMERALDS Trade Price D1')
ax1.set_ylabel('Price')

ax2.plot(tomatoes_trades_d1['timestamp'], tomatoes_trades_d1['price'], linewidth=0.8, color='red')
ax2.set_title('TOMATOES Trade Price D1')
ax2.set_ylabel('Price')
ax2.set_xlabel('Timestamp')

plt.tight_layout()
plt.show()
```
